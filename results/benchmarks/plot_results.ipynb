```python
import numpy as np
from matplotlib import pyplot as plt
data16 = np.load('method1_benchmark_16.npy')
data8 = np.load('method1_benchmark_8.npy')
data4 = np.load('method1_benchmark_4.npy')
data_b = np.load('../benchmarks/vanilla_benchmark.npy')
data3_4 = np.load('method1_benchmark_3_4.npz.npy')
data3_6 = np.load('method1_benchmark_3_6.npz.npy')
data3_8 = np.load('method1_benchmark_3_8.npz.npy')
data4_6 = np.load('method1_benchmark_4_6.npz.npy')

y0 = np.median([99.187911184210531,99.064555921052633,99.04485023413291])

y_dr = ([np.mean([0.978926809211, 0.9765625, 0.980057565789]),
      np.mean([0.973170230263, 0.974917763158,0.972553453947]),
      np.mean([0.970291940789,0.960834703947,0.968441611842])])

x1 = [np.mean(data4),np.mean(data8),np.mean(data16)]
x2 = [np.mean(data3_4),np.mean(data3_6),np.mean(data3_8),np.mean(data4_6)]


y = np.matrix.round(np.array(y_dr)*100,4)
y = y - y0
z = 100*np.array([0.99, 0.9795, 0.68657, 0.77569901315789469,0.84375,0.93883634868421051,0.95836759868421051])- y0
f = plt.figure()
f.add_subplot()
plt.title("Results, speed vs. loss")
plt.ylabel('accuracy loss, %')
plt.xlabel('time, sec')
data1, = plt.plot(x1, y, 'r^',label='data reconstruction')
data2, =plt.plot(x1+x2,z,'go',label='filter reconstruction')
data3, = plt.plot(np.median(data_b),0,'bs', label='original method')
plt.ylim((-4,0.1))
plt.legend(bbox_to_anchor=(1.0, 0.33))
plt.show()
```
